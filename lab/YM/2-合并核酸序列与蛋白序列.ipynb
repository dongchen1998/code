```python
import pandas as pd
```

```python
df_nucl = pd.read_csv('./Nucleic_acid_sequence.csv').copy()

df_nucl.rename(
    columns={"gene": "Gene", 
                    "protein_id":"Protein_id", 
                    'ncbi_gene_id': 'NCBI_Gene_ID',
                    "sequence": "Nucleic_acid_sequence"}, 
    inplace=True
)
df_nucl
```

```python
df_protein = pd.read_csv("./Amino_acid_sequence.csv").copy()

df_protein.rename(
    columns={"Protein ID": "Protein_id",
             "Sequence": "Amino_acid_sequence"},
    inplace=True,
)

df_protein
```

```python
# 合并核酸序列与蛋白序列
df = pd.merge(df_nucl, df_protein, on="Protein_id", how="outer")

# df重新排列
df = df[
    [
        "Gene",
        "Protein_id",
        'NCBI_Gene_ID',
        "Nucleic_acid_sequence",
        "Amino_acid_sequence",
    ]
]

df
```

```python
# 保存
df.to_csv('./Sc-S288C_Sequence.csv', index=False)
# df.to_csv('./Nc_OR74A_Sequence.csv', index=False)
```
