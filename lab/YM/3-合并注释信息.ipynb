```python
import pandas as pd

# 读取数据
df_seq = pd.read_csv('./Sc-S288C_Sequence.csv').copy()
df_seq
```

```python
df_ano = pd.read_csv('./Sc-S288C_Ano-raw.tsv',sep='\t').copy()
df_ano
```

```python
# df_ano的seed_ortholog列，每一行不保留前5个字符
df_ano['seed_ortholog'] = df_ano['seed_ortholog'].str[5:]
# df_ano的seed_ortholog列改名为gene_id
df_ano.rename(columns={'seed_ortholog':'Gene'},inplace=True)
df_ano.rename(columns={'#query':'Protein_id'},inplace=True)
df_ano.head()
```

```python
# df_ano重新排列列的顺序
df_ano = df_ano[['Gene', 'Description', 'Preferred_name',
                 'GOs', 'KEGG_ko', 'KEGG_Pathway', 'KEGG_Module', 'KEGG_Reaction','KEGG_rclass',
                 'EC', 'CAZy','PFAMs']]

df_ano.head()
```

```python
# 合并df_seq和df_ano，根据Gene列,合并方式为
df_merge = pd.merge(df_seq,df_ano,how='left', on='Gene')
df_merge
```

```python
# df_merge重新排列列的顺序
df_merge = df_merge[['Gene', 'Protein_id', 'NCBI_Gene_ID','Description', 'Preferred_name',
                 'GOs', 'KEGG_ko', 'KEGG_Pathway', 'KEGG_Module', 'KEGG_Reaction','KEGG_rclass',
                 'EC', 'CAZy','PFAMs','Nucleic_acid_sequence','Amino_acid_sequence']]

df_merge.head()
```

```python
# 保存为csv文件
df_merge.to_csv('./Sc-S288C_Ano.csv',index=False)
```
