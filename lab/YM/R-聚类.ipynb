```r
# Path: R.ipynb 读取Mt_Counts.csv
exp = read.csv("Mt_Counts.csv", header = TRUE, row.names = 1)

exp
```

```r
# exp不保留第二列到第六列
exp = exp[, -c(1:5)]
```

```r
# exp只保留前7列
exp = exp[, 1:6]
exp
```

```r
metadata <- data.frame(sample_id = colnames(exp))
metadata
```

```r
sample <- rep(c("con","treat"),each=3)
sample
```

```r
metadata$sample <- relevel(factor(sample),"con")
metadata
```

```r
library(DESeq2)
## 第一列有名称，所以tidy=TRUE
dds <-DESeqDataSetFromMatrix(countData=exp, 
                             colData=metadata, 
                             design=~sample
                             )
```

```r
# q: 为什么上诉代码报错？
# a: tidy=TRUE, 会报错，因为第一列有名称，所以tidy=TRUE
nrow(dds)
```

```r
dds <- dds[rowSums(counts(dds))>1,]
```

```r
nrow(dds)
```

```r
vsd <- vst(dds, blind = FALSE)
```

```r
sampleDists <- dist(t(assay(vsd)))
```

```r
hc <- hclust(sampleDists, method = "ward.D2")
```

```r
plot(hc, hang = -1)
```

```r
library(factoextra)
res <- hcut(sampleDists, k = 2, stand = TRUE)
# Visualize
fviz_dend(res, 
          # 加边框
          rect = TRUE,
          # 边框颜色
          rect_border="cluster",
          # 边框线条类型
          rect_lty=2,
          # 边框线条粗细
          lwd=1.2,
          # 边框填充
          rect_fill = T,
          # 字体大小
          cex = 1,
          # 字体颜色
          color_labels_by_k=T,
          # 平行放置
          horiz=T)
```
