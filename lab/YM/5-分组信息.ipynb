```python
import pandas as pd
import numpy as np

# 读取数据的sheet2
df = pd.read_excel('./Mt_sample.xlsx', sheet_name='Sheet2').copy()
df_count = pd.read_csv('./Mt_Counts_autohagy.csv').copy()
```

```python
# 假设df和df_count的内容如题目所示

# 提取df_count的第一列作为索引列
df_count_index = df_count[['Geneid']] 

# 提取df_count除第一列外的所有列,并设置columns
df_count_data = df_count.iloc[:,1:]
df_count_data.columns = df['sample_id']

# 合并索引列和排序后的数据列
df_count = pd.concat([df_count_index, df_count_data], axis=1)

df_count
```

```python
# 删除掉condition列中没有重复的行
df = df[df.duplicated(subset=['condition'], keep=False)]

# 保存为csv
df.to_csv('./Mt_sample_info.csv', index=False)
```

```python
# df_count从第二列开始，只保留df的sample_id中记录的列
df_count = df_count[['Geneid'] + list(df['sample_id'])]
df_count
```

```python
# 保存为csv
df_count.to_csv('./Mt_Counts_autohagy_deseq.csv', index=False)
```
