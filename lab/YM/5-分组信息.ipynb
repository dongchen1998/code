```python
import pandas as pd
import numpy as np
```

```python

# 读取数据的sheet2
df = pd.read_excel('./Nc_sample.xlsx', sheet_name='Sheet2').copy()
df_count = pd.read_csv('./Nc_Counts_autohagy.csv').copy()
```

```python
df['sample_id'] = df['sample_id'].apply(lambda x: x.split(','))
df = df.explode('sample_id').reset_index(drop=True)
df
```

```python
# df重新排列
df = df[['sample_id', 'condition']]
df
```

```python
# Select only the columns in df_count that are also in df['sample_id']
cols_to_select = ['Geneid'] + list(df['sample_id'])
df_count_selected = df_count[cols_to_select]

# Set the Geneid column as the index
df_count_selected.set_index('Geneid', inplace=True)

# Transpose the dataframe so that the sample_id values become the column names
df_count_selected = df_count_selected

# 重制索引
df_count_selected.reset_index(inplace=True)

df_count_selected
```
