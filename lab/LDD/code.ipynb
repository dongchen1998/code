```python
# 读取Lab/LDD/AJ304803.1.txt，将第二列的序列中的空格删除

import re

with open('AJ304803.1.txt', 'r') as f:
    seq = f.readlines()[1].strip()
    seq = re.sub(' ', '', seq)
    print(seq)

    # 将小写字母转换为大写字母
    seq = seq.upper()
    print(seq)

    # 保存到文件
    with open('AJ304803.1.txt', 'w') as f:
        f.write(seq)
```

```python
# 计算序列长度
print(len(seq))
```
