```python
import pandas as pd
```

```python
keyname = 'Nc'
path = '/Users/dongjiacheng/Desktop/coder/mtd/RNA-seq/'+ keyname +'/'
print(path)
```

1. 标准化

```python
df_sample = pd.read_excel(path + keyname + "_sample_raw_2023.12.28.xlsx").copy()

df_sample['Strategy'] = df_sample['Strategy'].fillna('')
df_sample['Carbon Sources'] = df_sample['Carbon Sources'].fillna('')
df_sample['Temperature'] = df_sample['Temperature'].fillna('')
df_sample['Time'] = df_sample['Time'].fillna('')

# Strain列转为str类型
df_sample['Strain'] = df_sample['Strain'].astype(str) 

df_sample['Carbon Sources'] = df_sample['Carbon Sources'].apply(lambda x: x.strip())
df_sample['Strain'] = df_sample['Strain'].apply(lambda x: x.strip())

df_sample['Strategy'] = df_sample['Strategy'].apply(lambda x: x.capitalize())
df_sample['Carbon Sources'] = df_sample['Carbon Sources'].apply(lambda x: x.capitalize())

df_sample
```

聚合
1. 一个fname号对应一个或者多个样本
2. 聚合后，则方便后续查看平均值

```python
df_sample_change = df_sample.groupby('Sample ID').agg({
    'Sample Name':lambda x: ','.join(x),
    'Experiment':lambda x: ','.join(x),
    'Run':lambda x: ','.join(x),
    'Series Accession':lambda x: ','.join(x),
    'Study Title':lambda x: ';'.join(x),
    'Species':lambda x: ','.join(x),
    'Model':lambda x: ','.join(x),
    'Library Seletion':lambda x: ','.join(x),
    'Library Source':lambda x: ','.join(x),
    'Library Strategy':lambda x: ','.join(x),
    'Samples':lambda x: ','.join(x),
    'Strain':lambda x: ','.join(str(i) for i in x),
    'Strategy':lambda x: ','.join(str(i) for i in x),
    'Carbon Sources':lambda x: ','.join(str(i) for i in x),
    'Temperature':lambda x: ','.join(str(i) for i in x),
    'Time':lambda x: ','.join(str(i) for i in x),
    'Year':lambda x: ','.join(str(i) for i in x),
    }).reset_index()

df_sample_change
```

```python
"""
Series ID这一列中，存在多个值，如果值相同则值合并
"""
df_sample_change['Series Accession'] = df_sample_change['Series Accession'].str.split(',').str[0]
df_sample_change['Study Title'] = df_sample_change['Study Title'].str.split(';').str[0]
df_sample_change['Species'] = df_sample_change['Species'].str.split(',').str[0]
df_sample_change['Strain'] = df_sample_change['Strain'].str.split(',').str[0]
df_sample_change['Model'] = df_sample_change['Model'].str.split(',').str[0]
df_sample_change['Library Seletion'] = df_sample_change['Library Seletion'].str.split(',').str[0]
df_sample_change['Library Source'] = df_sample_change['Library Source'].str.split(',').str[0]
df_sample_change['Library Strategy'] = df_sample_change['Library Strategy'].str.split(',').str[0]
df_sample_change['Strain'] = df_sample_change['Strain'].str.split(',').str[0]
df_sample_change['Strategy'] = df_sample_change['Strategy'].str.split(',').str[0]
df_sample_change['Carbon Sources'] = df_sample_change['Carbon Sources'].str.split(',').str[0]
df_sample_change['Temperature'] = df_sample_change['Temperature'].str.split(',').str[0]
df_sample_change['Time'] = df_sample_change['Time'].str.split(',').str[0]
df_sample_change['Year'] = df_sample_change['Year'].str.split(',').str[0]

df_sample_change.head()
```

```python
"""
筛选目标列
"""
df_sample_filter = df_sample_change[['Sample ID', 'Strain', 'Strategy' ,'Carbon Sources', 'Temperature', 'Time', 'Study Title', 'Series Accession', 'Sample Name']].copy()

df_sample_filter.rename(columns={'Sample Name':'Related Sample'}, inplace=True)

df_sample_filter
```

```python
df_sample_filter.to_csv(path + keyname + '_sample.tsv', sep='\t', index=False)
```
