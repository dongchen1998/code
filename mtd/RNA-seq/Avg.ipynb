```python
import pandas as pd

# 读取原始数据文件，读取第二张表
# df_1 = pd.read_excel('./An-CBS-513.88-FPKM.xlsx', sheet_name=1).copy()
# df_1 = pd.read_excel('./Counts-FPKM.xlsx', sheet_name=1).copy()
# df = pd.read_excel('An-CBS-513.88_公共数据.xlsx', sheet_name=1)
df = pd.read_excel('./LQ/LDD/An-5-FPKM.xlsx')
df
```

```python
# 三个样本求平均
group_columns = [df.columns[i:i+3] for i in range(1, len(df.columns), 3)]
for columns in group_columns:
    column_name = columns[0]
    column_name = column_name[:-2]
    column_name = column_name + '_avg '
    df[column_name] = df[columns].mean(axis=1).round(2)

# # 两个样本求平均
# group_columns = [df.columns[i:i+2] for i in range(1, len(df.columns), 2)]
# for columns in group_columns:
#     column_name = columns[0]
#     column_name = column_name[:-2]
#     column_name = column_name + '_avg '
#     df[column_name] = df[columns].mean(axis=1).round(2)
```
