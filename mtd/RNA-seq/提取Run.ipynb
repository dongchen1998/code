```python
import pandas as pd

df = pd.read_csv('./Nc_RPKM_Run.csv')
df
```

```python
# 将df所有的列名输出到一个txt文件中
# 获取所有列名
cols = list(df.columns)
# 写入txt文件
with open('sra.txt','w') as f:
    for col in cols:
        f.write(col + '\n')
```

```python
import os

path = './Counts/'

with open('sra.txt', 'w') as f:
    for file in os.listdir(path):
        if file.endswith('.txt'):
            f.write(file + '\n')
```
