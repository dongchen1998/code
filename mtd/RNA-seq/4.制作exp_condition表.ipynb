```python
import pandas as pd
```

```python
keyname = 'Nc'
path = '/Users/dongjiacheng/Desktop/coder/mtd/RNA-seq/'+ keyname +'/'
print(path)
```

```python
exp_df = pd.read_csv(path + keyname + '_exp.tsv', sep='\t').copy()
sample_df = pd.read_csv(path + keyname + '_sample.tsv', sep='\t').copy()

# 将Temperature空值设置为''
sample_df['Temperature'] = sample_df['Temperature'].fillna('')
sample_df['Carbon Sources'] = sample_df['Carbon Sources'].fillna('')
sample_df['Time'] = sample_df['Time'].fillna('')

# 将Sample ID、Strain、Carbon Sources、Temperature、Time的值组合成新的列名
sample_df['New_Column_Name'] = sample_df['Sample ID'] + ' ' + sample_df['Strain'] + ' ' + sample_df['Carbon Sources'] + ' ' + sample_df['Temperature'] + ' ' + sample_df['Time'].astype(str)

# 创建一个映射字典
rename_dict = sample_df.set_index('Sample ID')['New_Column_Name'].to_dict()

# 使用字典重命名exp.tsv的列名
exp_df.rename(columns=rename_dict, inplace=True)

# 查看更改后的exp.tsv
exp_df
```

```python
# 保存exp_df为tsv
exp_df.to_csv(path + keyname + '_exp_condition.tsv', sep='\t', index=False)
```
