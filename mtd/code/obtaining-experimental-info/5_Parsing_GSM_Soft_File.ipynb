### 解析GSM.soft文件

```python
import re
import os
import pandas as pd

def extract_information(file_path, df):
    """
    从GSM文件中提取特定信息，并将其存储在 DataFrame 中。
    """

    with open(file_path) as file:
        content = file.read()

    sample_accession = r'!Sample_geo_accession = (\w+)'
    try:
        sample = re.findall(sample_accession, content)
    except:
        sample = 'N/A'  

    organism = r'!Sample_organism_ch1 = (.*)'
    try:
        species = re.findall(organism, content)[0]
    except:
        species = 'N/A'

    strain_name = r'!Sample_characteristics_ch1 = (.*)'
    try:
        strain = re.findall(strain_name, content)[0]
    except: 
        strain = 'N/A'

    condition = r'!Sample_characteristics_ch1 = (.*)'
    try:
        condition = re.findall(condition, content)[1]
    except: 
        condition = 'N/A'

    condition_2 = r'!Sample_characteristics_ch1 = (.*)'
    try:
        condition_2 = re.findall(condition_2, content)[2]
    except: 
        condition_2 = 'N/A'
    
    condition_3 = r'!Sample_characteristics_ch1 = (.*)'
    try:
        condition_3 = re.findall(condition_2, content)[3]
    except: 
        condition_3 = 'N/A'

    taxid_id = r'!Sample_taxid_ch1 = (.*)'
    try:
        taxid = re.findall(taxid_id, content)[0]
    except:
        taxid = 'N/A'

    # protocol_txt = r'Sample_growth_protocol_ch1 = (.*)'
    # try:
    #     proticol = re.findall(protocol_txt, content)[0]
    # except: 
    #     proticol = 'N/A'

    model_id = r'Sample_instrument_model = (.*)'
    try:
        model = re.findall(model_id,content)
    except:
        model = 'N/A'

    selection_id = r'!Sample_library_selection = (.*)'
    try:
        library_selection = re.findall(selection_id,content)
    except:
        library_selection = 'N/A'

    source_id = r'!Sample_library_source = (.*)'
    try:
        library_source = re.findall(source_id,content)
    except:
        library_source = 'N/A'

    strategy_id = r'!Sample_library_strategy = (.*)'
    try:
        library_strategy = re.findall(strategy_id,content)
    except:
        library_strategy = 'N/A'

    spx_id = r'!Sample_relation = SRA:(.*)'
    try:
        experiment = re.findall(spx_id, content)
    except:
        experiment = 'N/A'

    series_id = r'!Sample_series_id = (.*)'
    try:
        series = re.findall(series_id,content)[0]
    except:
        series = 'N/A'


    # df_temp = pd.DataFrame(columns=['Sample','Species','Strain','TAXID','Protocol','Model','Library_seletion','Library_source','Library_strategy','Series','Condition','Condition_2','Condition_3'])
    df_temp = pd.DataFrame(columns=['Sample','Species','Strain','TAXID','Model','Library_seletion','Library_source','Library_strategy','Experiment','Series','Condition','Condition_2','Condition_3'])
    
    df_temp['Sample'] = sample
    df_temp['Species'] = species
    df_temp['Strain'] = strain
    df_temp['TAXID'] = taxid
    # df_temp['Protocol'] = proticol
    df_temp['Model'] = model
    df_temp['Library_seletion'] = library_selection
    df_temp['Library_source'] = library_source
    df_temp['Library_strategy'] = library_strategy
    df_temp['Experiment'] = experiment
    df_temp['Series'] = series
    df_temp['Condition'] = condition
    df_temp['Condition_2'] = condition_2
    df_temp['Condition_3'] = condition_3

    df = pd.concat([df, df_temp], ignore_index=True)

    return df
```

```python
path = './Nc/GSM_soft/'
key_name ='Nc'
```

```python
# df = pd.DataFrame(columns=['Sample','Species','Strain','TAXID','Protocol','Model','Library_seletion','Library_source','Library_strategy','Series','Condition','Condition_2','Condition_3'])
df = pd.DataFrame(columns=['Sample','Strain','TAXID','Model','Library_seletion','Library_source','Library_strategy','Experiment','Series','Condition','Condition_2','Condition_3'])

for file in os.listdir(path):
    try:
        file_path = os.path.join(path, file)
        df = extract_information(file_path, df)
    except Exception as e:
        print(f"An error occurred while processing file:\n{file_path}\nError message: {str(e)}")

df
```

```python
# df新建列，名为Fname,Run,Study Title,Sample Name,值为空
df['Sample ID'] = ''
df['Run'] = ''
df['Study Title'] = ''
df['Samples'] = ''
df['Treatment'] = ''
df['Year'] = ''

df.rename(columns={'Sample':'Sample Name'}, inplace=True)
df.rename(columns={'Series':'Series Accession'}, inplace=True)
df.rename(columns={'Library_seletion':'Library Seletion'}, inplace=True)
df.rename(columns={'Library_strategy':'Library Strategy'}, inplace=True)
df.rename(columns={'Library_source':'Library Source'}, inplace=True)

df['Experiment'] = df['Experiment'].str.replace('https://www.ncbi.nlm.nih.gov/sra?term=', '')

# Strain列中将source:改为空,strain:改为空,carbon source:改为空,tissue:改为空
df['Strain'] = df['Strain'].str.replace('source:', '')
df['Strain'] = df['Strain'].str.replace('strain:', '')
df['Strain'] = df['Strain'].str.replace('carbon source:', '')
df['Strain'] = df['Strain'].str.replace('tissue:', '')

# df重新排序
df = df[['Sample ID',
         'Sample Name',
         'Experiment',
         'Run',
         'Series Accession',
         'Study Title',
         'Species',
         'Strain',
         'Model',
         'Library Seletion',
         'Library Source',
         'Library Strategy',
         'Samples',
         'Treatment',
         'Condition',
         'Condition_2',
         'Condition_3',
         'Year']]

df.head()
```

```python
df_gse_info = pd.read_csv(key_name + '_dataset.csv').copy()

# df_gse_info不保存Title列
df_gse_info.drop(['Title'], axis=1, inplace=True)

df_gse_info
```

给Sample表添加Study Title

```python
df.loc[:, 'Study Title'] = df['Series Accession'].map(df_gse_info.set_index('Series Accession')['Study Title'])
df
```

给sample表添加samples名称

```python
df_name = pd.read_csv(key_name + '_GSM_ID.csv').copy()

# 添加Sample Name的内容
for i in range(len(df)):
    for j in range(len(df_name)):
        if df.loc[i, 'Sample Name'] == df_name.loc[j, 'Accession']:
            df.loc[i, 'Samples'] = df_name.loc[j, 'Title']

df
```

```python
for i in range(len(df)):
    for j in range(len(df_gse_info)):
        if df['Series Accession'][i] == df_gse_info['Series Accession'][j]:
            df.loc[i, 'Year'] = df_gse_info.loc[j, 'Year']

# df = df.sort_values(by='Year', ascending=False)
df
```

```python
# 删除掉Study Title为空的行
df_only_pub = df.dropna(subset=['Study Title'])

# 重置索引
df_only_pub = df_only_pub.reset_index(drop=True)

df_only_pub
```

```python
df_ncbi = pd.read_csv(key_name + '_SraRunInfo.csv').copy()
df_ncbi
```

```python
df_ncbi = df_ncbi[['Run', 'Experiment']].copy()

# 如果Experiment列存在重复，则将重复的行合并
df_ncbi = df_ncbi.groupby('Experiment').agg({
    'Run':lambda x: ','.join(x),
    }).reset_index().copy( )

df_ncbi
```

```python
# 保存为tsv
df_ncbi.to_csv (key_name + '_SraRunInfo_Run.csv', sep=',', index=False)
```

```python
df_nc_sample = df_only_pub.copy()

# 将空值填充为空
df_nc_sample = df_nc_sample.fillna('')

# 去掉Experiment列中的值的第一个空格
df_nc_sample['Experiment'] = df_nc_sample['Experiment'].apply(lambda x: x.strip())

df_nc_sample = df_nc_sample[df_nc_sample['Library Strategy'] == 'RNA-Seq']

df_nc_sample
```

```python
# 如果df_ncbi的Experiment列中的值与df_nc_sample的Experiment列中的值相同，则将df_ncbi的Run列中的值赋给df_nc_sample的Run
df_nc_sample['Run'] = df_nc_sample['Experiment'].map(df_ncbi.set_index('Experiment')['Run'])
df_nc_sample
```

```python
df = df_nc_sample.dropna(subset=['Run']) # 有些数据没Run，既没公开
df
```

```python
df.to_excel(key_name + '_sample.xlsx', index=False)
```
