```python
import pandas as pd


df1 = pd.read_csv('./Gene_Expression_Matrix.csv').copy()
df1
```

```python
df2 = pd.read_csv('./Mt_exp_condition.tsv', sep='\t').copy()
df2
```

```python
df1_columns = df1.columns.tolist()
df1_columns
```

```python
df2_columns = df2.columns.tolist()
df2_columns
```

```python
# Now let's create the mapping from df1 columns to df2 columns
# Assuming that the df2 contains all the column identifiers from df1 plus some additional information
mapping = {col: next((df2_col for df2_col in df2_columns if col in df2_col), None) for col in df1_columns}

# Apply the mapping to rename the columns in df1
df1_renamed = df1.rename(columns=mapping)

df1_renamed.columns # Display the new column names to verify
```

```python
# 将df1_renamed.columns作为df1的列名
df1.columns = df1_renamed.columns
df1
```

```python
# 保存为tsv文件
df1.to_csv('./Gene_Expression_Matrix_renamed.tsv', sep='\t', index=False)
```
