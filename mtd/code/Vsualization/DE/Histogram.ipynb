```python
import pandas as pd

# Load the uploaded CSV file
demo_fpkm_data = pd.read_csv("demo-fpkm-all.csv")

# Display the first few rows of the dataset
demo_fpkm_data.head()
```

```python
import plotly.graph_objects as go

# Extract all sample columns for gene expression distribution
sample_columns = demo_fpkm_data.columns[1:]
data = []

# For each sample, generate a histogram trace
for sample in sample_columns:
    data.append(
        go.Histogram(
            x=demo_fpkm_data[sample],
            opacity=0.75,
            name=sample,
            histnorm='percent'
        )
    )

# Layout configuration for the histogram
layout = go.Layout(
    title="Gene Expression Distribution across Samples",
    xaxis=dict(title="FPKM value"),
    yaxis=dict(title="Percentage of Genes"),
    barmode='overlay'
)

# Create the figure
fig = go.Figure(data=data, layout=layout)

fig
```
